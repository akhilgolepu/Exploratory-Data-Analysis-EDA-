Renaming Columns




```python
df2 = df2.rename(columns = {'Year of event': 'year',
                            'Event dates' : 'event_day',
                            'Event name' : 'event_name',
                            'Event distance/length' : 'event_distance',
                            'Event number of finishers' : 'event_finishers',
                            'Athlete performance' : 'athlete_performance',
                            'Athlete gender' : 'athlete_gender',
                            'Athlete average speed' : 'athlete_average_speed',
                            'Athlete ID' : 'athlete_id'
                            })
```

Column Reordering

```python
df3 = df2[['event_day', 'event_name', 'event_distance', 'event_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed', 'year']]
```

Difference in speeds for the 50km, 50mi male to female

```python
df3.groupby(['event_distance', 'athlete_gender'])['athlete_average_speed'].mean()
```

What age groups are the best in the 50mi race (20 + races min)

```python
df3.query('event_distance == "50mi"').groupby('athlete_age')['athlete_average_speed'].aggregate(['mean', 'count']).sort_values(by='mean', ascending=False).query('count>19')
```

What age groups are the best in the 50mi race (20 + races min) (show 20)

```python
df3.query('event_distance == "50mi"').groupby('athlete_age')['athlete_average_speed'].aggregate(['mean', 'count']).sort_values(by='mean', ascending=False).query('count>19').head(20)
```

```python
df3['event_month'] = df3['event_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head(25)
```

```python
df3['event_season'] = df3['event_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring')
```

```python
df3.head(25)
```

```python
df3.groupby('event_season')['athlete_average_speed'].aggregate(['mean', 'count']).sort_values(by='mean', ascending=False)
```
